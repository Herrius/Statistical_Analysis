*Observaciones:*
- age: La mayoría de los estudiantes tienen entre 15 y 18 años, con una distribución ligeramente sesgada a la derecha.

- failures: La mayoría de los estudiantes no han reprobado ninguna asignatura en el pasado.

- goout: La mayoría de los estudiantes salen con amigos con una frecuencia moderada.

- health: La mayoría de los estudiantes reportan tener un buen estado de salud.
- absences: La mayoría de los estudiantes tienen pocas ausencias.
- G1, G2, G3: Las calificaciones en los tres periodos parecen tener una distribución aproximadamente normal, con la mayoría de los estudiantes obteniendo calificaciones en el rango medio.

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score

# Definir las variables predictoras y la variable objetivo
predictors = ['G1', 'Medu', 'studytime', 'failures', 'absences', 'goout']  # Ajusta las variables según sea necesario
target = 'G3'
```

```python
# Crear el modelo de regresión para matemáticas (df_mat)
X_mat = df_mat[predictors]
y_mat = df_mat[target]
X_train_mat, X_test_mat, y_train_mat, y_test_mat = train_test_split(X_mat, y_mat, test_size=0.2, random_state=42)
model_mat = LinearRegression()
model_mat.fit(X_train_mat, y_train_mat)
```

```python
# Crear el modelo de regresión para portugués (df_por)
X_por = df_por[predictors]
y_por = df_por[target]
X_train_por, X_test_por, y_train_por, y_test_por = train_test_split(X_por, y_por, test_size=0.2, random_state=42)
model_por = LinearRegression()
model_por.fit(X_train_por, y_train_por)
```

```python
# Evaluar los modelos
y_pred_mat = model_mat.predict(X_test_mat)
y_pred_por = model_por.predict(X_test_por)
```

```python
from sklearn.metrics import root_mean_squared_error
print("----- Resultados del modelo para Matemáticas (Mat) -----")
print(f"R-cuadrado: {r2_score(y_test_mat, y_pred_mat)}")
print(f"RMSE: {root_mean_squared_error(y_test_mat, y_pred_mat)}")
```

```python
print("\n----- Resultados del modelo para Portugués (Por) -----")
print(f"R-cuadrado: {r2_score(y_test_por, y_pred_por)}")
print(f"RMSE: {root_mean_squared_error(y_test_por, y_pred_por)}")
```
